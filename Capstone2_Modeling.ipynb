Conventional medicine is a system designed for a brief consultation with a medical doctor, typically lasting 15 minutes. The goal is to diagnose as quickly as possible and to treat ailments or diseases with prescribed pharmaceuticals. However, this approach often fails to address the root causes of systemic issues affecting a patient's whole body.

Conventional medicine struggles to address the public health crises facing Americans. It seeks to improve all health markers of a patient's bodily systems through protocols involving nutraceuticals, diet, nutrition, health tests, lifestyle therapies, and natural treatments, allowing a patient's body to heal itself completely or to its fullest extent. Naturopathic Medicine "was founded in the United States in 1901 by Benedict Lust"<a href="#mla_8"> [8]</a>, taken from a movement that was happening at the turn of the century in Europe.

**Differentially expressed genes (DEGs):** Refer to the process used in bioinformatics and genomics to discover genes that show significant differences in expression levels between different conditions, groups, or treatments. For example, DEGs could be genes that are upregulated (increased expression) or downregulated (decreased expression) in cancerous tissue compared to normal tissue<a href="#mla_133"> [133]</a>.


<br><blockquote style="font-style: italic; margin: 20px; border-left: 4px solid #ccc; padding-left: 15px;">Disease and health both have differentially expressed genes, especially in Alzheimer's<a href="#mla_134"> [134]</a>.</blockquote><br>


**Microarray Experiment:** A microarray experiment uses technology to measure the expression levels of many genes simultaneously. There are thousands of microscopic spots in a microchip array, that are filled with a specific gene sequence that is being searched for expression levels<a href="#mla_135"> [135]</a>.

**Gene Ontology:** is a comprehensive and standardized system that categorizes gene functions into biological processes, molecular functions, and cellular components. It allows researchers to annotate genes consistently, compare gene functions across species, and analyze biological data in a systematic way, leading to a deeper understanding of gene roles in various biological contexts<a href="#mla_136"> [136]</a>.

**Protein-Protein Interaction (PPI) Network:** are a graphical representation of the physical and functional interactions between proteins within a cell or organism<a href="#mla_137"> [137]</a>.

**Hub genes:** are genes that play central roles in biological networks, often characterized by having a large number of connections (high degree) within a network<a href="#mla_138"> [138]</a>.

**Brain-derived neurotrophic Factor (BDNF):** is a protein that plays a critical role in neuronal survival, neurogenesis, synaptic plasticity, and cognitive functions. Without BDNF neurodegenerative diseases like Alzheimer's disease, Parkinson's disease, and Huntington's disease may contribute to the progression of these conditions by impairing neuronal survival and function. It is one of the most important neurotrophins, a family of proteins that promote the survival, development, and function of neurons<a href="#mla_139"> [139]</a>.

**WW domain-containing transcription regulator protein 1 (WWTR1):** is a protein domain found in many proteins involved in cellular signaling and regulation. It mediates protein-protein interactions and plays a role in cellular processes such as cell growth, apoptosis, and differentiation. It is key to keep appropriate organ size, and plays a role in cancer<a href="#mla_140"> [140]</a>.

**Braak stages:** refer to a system used to describe the progression of Alzheimer's disease pathology in the brain, specifically focusing on the distribution of neurofibrillary tangles, which are abnormal aggregates of tau protein. This staging system helps in understanding the progression of the disease and correlating it with clinical symptoms and cognitive decline<a href="#mla_141"> [141]</a>.

- Braak Stage I: Early stage with minimal cognitive impairment, often asymptomatic.
- Braak Stage II: Mild cognitive impairment may begin to be noticeable, particularly in areas related to memory.
- Braak Stage III: Progressive cognitive decline is evident, with increased memory problems and other symptoms affecting daily life.
- Braak Stage IV: More pronounced cognitive deficits, including difficulties with language, spatial orientation, and executive functions.
- Braak Stage V: Severe cognitive impairment, with significant loss of independence and function. More extensive involvement of various cognitive domains.
- Braak Stage VI: Advanced stage of Alzheimer's disease with profound cognitive impairment and severe functional decline. Patients may experience significant difficulties with basic activities of daily living.


**Amyloid-beta (Aβ) 42 & β-secretase activity:** is a specific form of amyloid-beta peptide that is 42 amino acids long. It is one of the most studied peptides in Alzheimer's disease research. Aβ 42 is prone to aggregation and forms amyloid plaques, which are a hallmark of Alzheimer's disease. These plaques are toxic to neurons and are associated with cognitive decline. Elevated levels of Aβ 42 in cerebrospinal fluid (CSF) are often used as a biomarker for Alzheimer's disease. Decreased levels of Aβ 42 in CSF can indicate its increased deposition in the brain, which is characteristic of the disease. β-Secretase activity leads to the production of Aβ peptides, including Aβ 42. Increased β-secretase activity results in higher levels of Aβ peptides, which can contribute to the formation of amyloid plaques in the brain<a href="#mla_142"> [142]</a><a href="#mla_143"> [143]</a>.

```python
from scipy import stats

# Define columns for each group
dss_columns = ['DSS_6', 'DSS_5', 'DSS_4', 'DSS_3', 'DSS_2', 'DSS_1']
m_columns = ['M_5', 'M_4', 'M_3', 'M_2', 'M_1']
con_columns = ['Con_4', 'Con_3', 'Con_2', 'Con_1']

# Calculate means for each group
dss_means = ds_1[dss_columns].mean(axis=1)
m_means = ds_1[m_columns].mean(axis=1)
con_means = ds_1[con_columns].mean(axis=1)

# Combine into a new DataFrame for easier analysis
df_means = pd.DataFrame({
    'DSS': dss_means,
    'M': m_means,
    'Control': con_means
})

# 1. **Descriptive Statistics**
print("Descriptive Statistics:")
print(df_means.describe())

# 2. **Statistical Test Example: t-Test Between Groups**
# Null Hypothesis: The means of 'DSS' and 'M' are equal
t_stat, p_val = stats.ttest_ind(df_means['DSS'], df_means['M'])
print("\nT-test between DSS and M:")
print(f"T-statistic: {t_stat}, P-value: {p_val}")

# Example: t-Test between DSS and Control
t_stat, p_val = stats.ttest_ind(df_means['DSS'], df_means['Control'])
print("\nT-test between DSS and Control:")
print(f"T-statistic: {t_stat}, P-value: {p_val}")

# Example: ANOVA to compare all groups
f_stat, p_val = stats.f_oneway(df_means['DSS'], df_means['M'], df_means['Control'])
print("\nANOVA between DSS, M, and Control:")
print(f"F-statistic: {f_stat}, P-value: {p_val}")

# 3. **Visualizations**
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to visualize the distributions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_means)
plt.title('Distribution of Means for DSS, M, and Control')
plt.show()



# Consistency Check
 #Observed vs. Permutation Results: The permutation test results suggested a significant difference, and the descriptive statistics and t-tests align with this finding. The means of the groups show that the DSS group has a higher mean than the Control but lower than the M group. This aligns with your permutation test result, which indicated that the observed difference was much larger than what would be expected by chance alone.
 # T-tests and ANOVA: The t-tests and ANOVA also show significant differences between groups. This supports the permutation test finding that there are real differences between the groups. The p-values from the t-tests are well below common significance levels, reinforcing the idea that the differences are statistically significant.
 #Conclusion
 #The permutation test's p-value of 0.0 and the statistical tests (t-tests and ANOVA) all indicate significant differences between the DSS group and the M+Control groups. This consistency between different methods supports the conclusion that the observed differences are significant and not due to random chance. -->
```

```python
import numpy as np

# Define your columns
dss_columns = ['DSS_6', 'DSS_5', 'DSS_4', 'DSS_3', 'DSS_1']
m_columns = ['M_5', 'M_4', 'M_3', 'M_2', 'M_1']
con_columns = ['Con_4', 'Con_3', 'Con_2', 'Con_1']

# Extract the data
dss_values = ds_1[dss_columns].values.flatten()
m_con_values = ds_1[m_columns + con_columns].values.flatten()

# Compute the observed difference
mean_dss = np.mean(dss_values)
mean_m_con = np.mean(m_con_values)
observed_diff = mean_dss - mean_m_con

print("Observed Difference:", observed_diff)

# Perform permutations
num_permutations = 10000
permutation_differences = []
combined_values = np.concatenate([dss_values, m_con_values])

for i in range(num_permutations):
    np.random.shuffle(combined_values)
    
    # Split shuffled data
    new_dss_values = combined_values[:len(dss_values)]
    new_m_con_values = combined_values[len(dss_values):]
    
    # Compute means
    new_dss_mean = np.mean(new_dss_values)
    new_m_con_mean = np.mean(new_m_con_values)
    
    # Compute permutation difference
    perm_diff = new_dss_mean - new_m_con_mean
    permutation_differences.append(perm_diff)
    
    # Print sample values for debugging
    if i % 1000 == 0:  # Print every 1000 permutations
        print(f"Permutation {i}:")
        print(f"New DSS Mean: {new_dss_mean}")
        print(f"New M+Con Mean: {new_m_con_mean}")
        print(f"Permutation Difference: {perm_diff}")
        print(f"Length of New DSS Values: {len(new_dss_values)}")
        print(f"Length of New M+Con Values: {len(new_m_con_values)}")
        print("-" * 40)

permutation_differences = np.array(permutation_differences)

# Calculate p-value
p_value = np.mean(np.abs(permutation_differences) >= np.abs(observed_diff))

# Print results
print('Observed Difference Mean:', observed_diff)
print("Permutation Difference Mean:", np.mean(permutation_differences))
print("P-value:", p_value)
```
